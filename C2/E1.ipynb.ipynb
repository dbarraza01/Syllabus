```python
#Ejercicio Formativo 1 Capítulo 2 Diego Barraza
import pandas as pd
from google.colab import files

uploaded = files.upload()
for filename in uploaded.keys():
    print('Archivo subido:', filename)

datos = pd.read_csv(filename)
print(datos)
```

```python
media = datos.groupby('Set').mean()
print(media)

desviacion_estandar = datos.groupby('Set').std()
print(desviacion_estandar)

maximo = datos.groupby('Set').max()
print(maximo)

minimo = datos.groupby('Set').min()
print(minimo)

correlacion = datos['X'].corr(datos['Y'])
print(correlacion)

kurtosisx = datos['X'].kurtosis()
print(kurtosisx)

```
