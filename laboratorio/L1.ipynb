```python
import datetime
import random
class Invernadero:
    def __init__(self,luz,temperatura):
        self.luz = 0
        self.temperatura = 0

    def luz_exterior(self):
        hora_actual = datetime.datetime.now().time()
        if hora_actual >= datetime.time(6, 0) and hora_actual <= datetime.time(12, 0):
            self.luz = (hora_actual.hour - 6) * 100 / 6
        elif hora_actual > datetime.time(12, 0) and hora_actual <= datetime.time(19, 0):
            self.luz = (12 - (hora_actual.hour - 12)) * 100 / 7
        else:
            self.luz = 0

    def temperatura_exterior(self):
        luz_sensor = Luz(100)
        luz_exterior = luz_sensor.medir()
        aleatorio = random.randint(-10, 10)
        hora_actual = datetime.datetime.now().time()
        if hora_actual >= datetime.time(6, 0) and hora_actual <= datetime.time(19, 0):
            self.temperatura = luz_exterior * 3 / 10 + aleatorio
        else:
            self.temperatura = 0

class Planta:
    def __init__(self,fecha_plantacion, humedad_optima):
        self.fecha_plantacion = fecha_plantacion #variable de tipo datetime
        self.humedad_optima = humedad_optima #variable de tipo numerico
        self.humedad_actual = 0

    def cosechar(self):
        pass

class Lechuga(Planta):
    def __init__(self,fecha_plantacion, humedad_optima, n_hojas, tamano):
        super().__init__(fecha_plantacion,humedad_optima)
        self.n_hojas = n_hojas #numero de hojas
        self.tamano = tamano

    def cosechar(self,cantidad):
        if cantidad <= self.n_hojas:
            self.n_hojas = self.n_hojas - cantidad
            if self.n_hojas > 40:
                self.tamano = "grande"
            elif self.n_hojas <= 40 and self.n_hojas >= 10:
                self.tamano = "mediana"
            elif self.n_hojas < 10:
                self.tamano = "chica"
    def crecer(self):
        # La lechuga crecerá una hoja cada vez que se realice un riego
        self.n_hojas += 1
        if self.n_hojas < 10:
            self.tamano = 'chica'
        elif 10 <= self.n_hojas < 40:
            self.tamano = 'mediana'
        else:
            self.tamano = 'grande'

class Tomate(Planta):
    def __init__(self, fecha_plantacion, humedad_optima, raza, n_tomates, acidez):
        super().__init__(fecha_plantacion, humedad_optima)
        self.raza = raza # normal o cherry
        self.n_tomates = n_tomates # numero de tomates de la planta
        self.acidez = acidez # % de acidez

    def cosechar(self, n_extraido):
        hora_actual = datetime.datetime.now()
        if self.raza == "cherry" and hora_actual < 12:
            return "no se pueden extraer tomates cherry antes de medio día"
        else:
            if n_extraido <= self.n_tomates:
                self.n_tomates -= n_extraido
                return f"Se ha cosechado {n_extraido} tomates"
            else:
                return "no hay suficientes tomates"

    def crecer(self):
        # El tomate crecerá un tomate cada vez que se realice un riego
        self.n_tomates += 1

class Cilantro(Planta):
    def __init__(self, fecha_plantacion, humedad_optima,n_ramas):
        super().__init__(fecha_plantacion, humedad_optima)
        self.n_ramas = n_ramas #numero de ramas

    def cosechar(self, n_extraido1):
        if n_extraido1 > self.n_ramas:
            return "No se puede extraer ese numero de ramas"
        elif n_extraido1 <= self.n_ramas:
            self.n_ramas -= n_extraido1
    def crecer(self):
        # El cilantro crecerá una rama cada vez que se realice un riego
        self.n_ramas += 1

class Zanahorias(Planta):
    def __init__(self, fecha_plantacion, humedad_optima, n_zanahorias):
        super().__init__(fecha_plantacion, humedad_optima)
        self.n_zanahorias = n_zanahorias

    def cosechar(self, n_extraido2):
        if n_extraido2 > self.n_zanahorias:
            return "No se puede extraer ese numero de zanahorias"
        elif n_extraido2 <= self.n_zanahorias:
            self.n_zanahorias -= n_extraido2
    def crecer(self):
        self.n_zanahorias +=1

class Sensores:
    def __init__(self):
        self.ultima_medicion = 0
    def medir(self):
        pass

class Humedad(Sensores):
    def medir(self, cantidad_plantas):
        # Simulación de medición de humedad
        disminucion = random.randint(1, 30)
        self.ultima_medicion = max(0, self.ultima_medicion - disminucion * cantidad_plantas)
        return self.ultima_medicion

class Temperatura(Sensores):
    def medir(self,temperatura_exterior, num_plantas):
        aleatorio = random.randint(1, 5)
        factor = num_plantas/10
        aleatorio *= factor
        if self.ultima_medicion < temperatura_exterior:
            self.ultima_medicion += aleatorio
            return self.ultima_medicion
        elif self.ultima_medicion > temperatura_exterior:
            self.ultima_medicion -= aleatorio
            return self.ultima_medicion
        elif self.ultima_medicion == temperatura_exterior:
            self.ultima_medicion = self.ultima_medicion
            return self.ultima_medicion

class Luz(Sensores):
    def medir(self, luz_exterior):
        return max(0, luz_exterior - random.randint(5, 10))

class Regador:
    def __init__(self, luz, temperatura, humedad):
        self.luz = luz
        self.temperatura = temperatura
        self.humedad = humedad

    def regar(self, planta):
        if self.luz > 60 and self.temperatura > 10 and self.humedad < 12:
            cantidad_agua_disponible = 100
            if cantidad_agua_disponible > 0:
                pass
            else:
                print("No hay suficiente agua disponible para regar.")
        else:
            print("No se cumplen las condiciones para regar la planta.")
```
